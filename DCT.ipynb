DCT for Table1

```python
from scipy.fftpack import dct
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('Table1.txt', 'r') as f:
    lines = f.readlines()
```

```python
data = np.array([line.strip().split() for line in lines], dtype=float)
```

```python
dct_data = dct(data, type=2, axis=1, norm='ortho')
```

```python
explained_variance_ratio = np.var(dct_data, axis=0) / np.sum(np.var(dct_data, axis=0))
cumulative_explained_variance_ratio = np.cumsum(explained_variance_ratio)
```

```python
plt.plot(cumulative_explained_variance_ratio)
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('DCT for Optimal Number of Components')
plt.grid(True)
plt.show()
```

```python
optimal_num_components = 35
```

```python
reduced_data_dct = dct_data[:, :optimal_num_components]
```

```python
print(reduced_data_dct.shape[1])
```
